## Função Constante

A função constante é uma das funções mais simples em matemática. Ela é definida como uma função que retorna sempre o mesmo valor, independentemente do valor da variável de entrada.

### Fórmula

A função constante é definida pela seguinte fórmula:

$$f(x) = c$$

onde:

$x$ é a variável de entrada
$c$ é o valor constante
### Exemplo Prático Manual

Vamos considerar a função constante $f(x) = 5$. Queremos calcular o valor da função para diferentes valores de $x$.

### Passo a Passo

Calcule o valor da função para $x = 2$:
$$f(2) = 5$$

Calcule o valor da função para $x = 5$:
$$f(5) = 5$$

Calcule o valor da função para $x = 10$:
$$f(10) = 5$$

Observe que, independentemente do valor de $x$, a função sempre retorna o valor $5$.

> Resolução em Julia

Aqui está o código em Julia para definir e calcular a função constante:

```julia
using Plots
theme(:dracula)

# Defina a função constante
c = 5
f(x) = c

# Calcule o valor da função para diferentes valores de x
x_values = [2, 5, 10]
for x in x_values
    println("f($x) = $(f(x))")
end

# Gera os valores de x para o plot
x_range = 0:0.1:12
y_values = f.(x_range)

# Plota a função constante
plot(x_range, y_values, label="f(x) = 5", legend=:bottomright, linewidth=2, color=:blue)
scatter!(x_values, f.(x_values), label="Pontos calculados", color=:red)
```

# Função Quadrática

A **função quadrática** (ou **função polinomial do 2º grau**) é uma função $ f: \mathbb{R} \rightarrow \mathbb{R} $ definida por:

$$
f(x) = ax^2 + bx + c
$$

Onde:
- $ a, b, c $ são coeficientes reais ($ a \neq 0 $).
- O gráfico é uma **parábola**.

---

## Principais Características

### 1. **Concavidade**
- Se $ a > 0 $, a parábola tem concavidade para **cima** (possui ponto de mínimo).
- Se $ a < 0 $, a parábola tem concavidade para **baixo** (possui ponto de máximo).

### 2. **Zeros (Raízes)**
Soluções de $ f(x) = 0 $, dadas pela **Fórmula de Bhaskara**:

$$
x = \frac{-b \pm \sqrt{\Delta}}{2a}, \quad \text{onde} \quad \Delta = b^2 - 4ac
$$

- **Δ > 0**: Duas raízes reais distintas.
- **Δ = 0**: Uma raiz real dupla.
- **Δ < 0**: Nenhuma raiz real (as raízes são complexas).

### 3. **Vértice da Parábola**
Ponto de máximo/mínimo da função:

$$
V = \left( -\frac{b}{2a}, -\frac{\Delta}{4a} \right)
$$

### 4. **Eixo de Simetria**
Reta vertical que passa pelo vértice:

$$
x = -\frac{b}{2a}
$$

---

## Exemplo Prático
Considere $ f(x) = x^2 - 4x + 3 $:
- **Coeficientes**: $ a = 1 $, $ b = -4 $, $ c = 3 $.
- **Concavidade**: Para cima ($ a > 0 $).
- **Raízes**: $ x = 1 $ e $ x = 3 $ (Δ = 4 > 0).
- **Vértice**: $ V = (2, -1) $.

- Cálculo de trajetórias (lançamento de projéteis).
- Otimização (custos, áreas, lucros).
- Análise de sinais em inequações.

### O que é uma Parábola?

É uma **curva simétrica** no plano cartesiano, com formato de "U" (ou "∩"), definida por todos os pontos $(x, y)$ que satisfazem a equação:

$$
y = ax^2 + bx + c
$$

### Características principais:

### 1. Formato
- Se $a > 0$: Parábola abre para **cima** (como um "U").
- Se $a < 0$: Parábola abre para **baixo** (como um "∩").

### 2. Elementos-chave
- **Vértice**: Ponto mais alto (se $a < 0$) ou mais baixo (se $a > 0$) da curva. É o "centro" da parábola.
- **Eixo de simetria**: Linha vertical que passa pelo vértice, dividindo a parábola em duas partes iguais e simétricas.
- **Raízes (zeros)**: Pontos onde a parábola cruza o eixo $x$ (se existirem).

### 3. Propriedades geométricas
- **Simetria**: Qualquer ponto $(x, y)$ na parábola tem um "espelho" $(-x + h, y)$ em relação ao eixo de simetria (onde $h$ é a coordenada $x$ do vértice).
- **Diretriz e Foco**: Em geometria analítica, uma parábola é definida como o conjunto de pontos equidistantes de:
  - Um ponto fixo (**foco**)
  - Uma reta fixa (**diretriz**)
  
  Na função quadrática, o foco e a diretriz estão implícitos nos coeficientes $a$, $b$, $c$.

> Curiosidade
A palavra "parábola" vem do grego parabolé (comparação), mas na matemática, foi adaptada por Apolônio de Perga (século III a.C.) para descrever essa curva, justamente por sua propriedade de "comparar" distâncias (foco e diretriz).

```julia
using Plots

# Define a função quadrática
f(x) = 2x^2 - 4x + 1

# Intervalo de valores para x (de -2 a 4, com passo 0.1)
x = -2:0.1:4

# Calcula os valores de y = f(x)
y = f.(x)

# Configurações do gráfico
plot(x, y, 
    label = "f(x) = 2x² - 4x + 1",  # Legenda
    title = "Parábola em Julia",     # Título
    xlabel = "Eixo x",               # Rótulo do eixo x
    ylabel = "Eixo y",               # Rótulo do eixo y
    linewidth = 2,                   # Espessura da linha
    color = :blue,                   # Cor da curva
    legend = :topleft,               # Posição da legenda
    grid = true                      # Grade habilitada
)

# Adiciona o vértice (calculado analiticamente)
a, b, c = 2, -4, 1
x_vertice = -b / (2a)
y_vertice = f(x_vertice)
scatter!([x_vertice], [y_vertice], label = "Vértice", color = :red)
```

# Função Afim (Função Polinomial do 1º Grau)

## Definição
Uma função $f: \mathbb{R} \rightarrow \mathbb{R}$ é chamada de **função afim** quando pode ser escrita na forma:

$$
f(x) = ax + b
$$

Onde:
- $a$ é o **coeficiente angular** (taxa de variação)
- $b$ é o **coeficiente linear** (valor inicial/interseção com o eixo y)
- $a, b \in \mathbb{R}$ e $a \neq 0$ (se $a = 0$, torna-se uma função constante)

## Propriedades Principais

### 1. Gráfico
- **Sempre uma reta** inclinada
- Cruza o eixo y no ponto $(0, b)$
- Cruza o eixo x no ponto $(-\frac{b}{a}, 0)$ (raiz da função)

### 2. Coeficientes
| Coeficiente | Efeito no Gráfico | Exemplo |
|-------------|-------------------|---------|
| $a > 0$     | Reta crescente     | $2x + 1$ |
| $a < 0$     | Reta decrescente   | $-x + 3$ |
| $b$         | Deslocamento vertical | $x + 2$ (sobe 2 unidades) |

### 3. Taxa de Variação
- $a = \frac{\Delta y}{\Delta x} = \frac{f(x_2) - f(x_1)}{x_2 - x_1}$
- Representa quanto $y$ muda para cada unidade de $x$

## Exemplo Prático
Considere $f(x) = 3x - 2$:
1. **Coeficientes**: $a = 3$, $b = -2$
2. **Raiz**: $x = \frac{2}{3}$ (quando $f(x) = 0$)
3. **Interseção com eixo y**: $(0, -2)$

> Resolução em Julia
Aqui está o código em Julia para definir e calcular a função afim:

```julia
# Defina a função afim
a = 2
b = 3
f(x) = a*x + b

# Calcule o valor da função para diferentes valores de x
x_values = [0, 1, 2]
for x in x_values
    println("f($x) = $(f(x))")
end

# Plote a função
using Plots
x = -10:0.1:10
y = f.(x)
plot(x, y, label="f(x) = 2x + 3", title="Função Afim")
```

Função Cúbica
A função cúbica é uma das funções mais importantes em matemática. Ela é definida como uma função que pode ser escrita na forma:

$$f(x) = ax^3 + bx^2 + cx + d$$

onde:

$x$ é a variável de entrada
$a$, $b$, $c$ e $d$ são constantes
Fórmula
A função cúbica pode ser escrita na forma:

$$f(x) = a(x - h)^3 + k$$

onde:

$h$ é o valor de $x$ que faz a função atingir seu mínimo ou máximo
$k$ é o valor da função no ponto $h$
Exemplo Prático Manual
Vamos considerar a função cúbica $f(x) = x^3 - 2x^2 + x - 1$. Queremos calcular o valor da função para diferentes valores de $x$.

Passo a Passo
Calcule o valor da função para $x = 0$:
$$f(0) = 0^3 - 2(0)^2 + 0 - 1 = -1$$

Calcule o valor da função para $x = 1$:
$$f(1) = 1^3 - 2(1)^2 + 1 - 1 = 0$$

Calcule o valor da função para $x = 2$:
$$f(2) = 2^3 - 2(2)^2 + 2 - 1 = 3$$

Observe que a função cúbica tem uma inclinação que muda de acordo com o valor de $x$.

Resolução em Julia
Aqui está o código em Julia para definir e calcular a função cúbica:

```julia
# Defina a função cúbica
a = 1
b = -2
c = 1
d = -1
f(x) = a*x^3 + b*x^2 + c*x + d

# Calcule o valor da função para diferentes valores de x
x_values = [0, 1, 2]
for x in x_values
    println("f($x) = $(f(x))")
end

# Plote a função
using Plots
x = -10:0.1:10
y = f.(x)
plot(x, y, label="f(x) = x^3 - 2x^2 + x - 1", title="Função Cúbica")
```

### Propriedades da Função Cúbica
A função cúbica tem algumas propriedades importantes:

- A função é contínua em todos os pontos do domínio.
- A função é diferenciável em todos os pontos do domínio, e a derivada é sempre uma função quadrática.
- A função é integrável em todos os intervalos do domínio.

Essas propriedades fazem com que a função cúbica seja uma ferramenta útil em muitas aplicações matemáticas e científicas.

## Função Polinomial
Uma função polinomial é uma função que pode ser escrita na forma:

$$f(x) = a_nx^n + a_{n-1}x^{n-1} + \cdots + a_1x + a_0$$

onde:

$x$ é a variável de entrada
$a_n, a_{n-1}, \ldots, a_1, a_0$ são constantes
$n$ é o grau da função polinomial
### Exemplo Prático Manual
Vamos considerar a função polinomial $f(x) = 2x^3 - 3x^2 + x - 1$. Queremos calcular o valor da função para diferentes valores de $x$.

### Passo a Passo
Calcule o valor da função para $x = 0$:
$$f(0) = 2(0)^3 - 3(0)^2 + 0 - 1 = -1$$

Calcule o valor da função para $x = 1$:
$$f(1) = 2(1)^3 - 3(1)^2 + 1 - 1 = 0$$

Calcule o valor da função para $x = 2$:
$$f(2) = 2(2)^3 - 3(2)^2 + 2 - 1 = 11$$

Observe que a função polinomial tem uma inclinação que muda de acordo com o valor de $x$.

> Resolução em Julia
Aqui está o código em Julia para definir e calcular a função polinomial:

```julia
# Defina a função polinomial
a = [2, -3, 1, -1]
f(x) = sum(a[i]*x^(length(a)-i) for i in 1:length(a))

# Calcule o valor da função para diferentes valores de x
x_values = [0, 1, 2]
for x in x_values
    println("f($x) = $(f(x))")
end

# Plote a função
using Plots
x = -10:0.1:10
y = f.(x)
plot(x, y, label="f(x) = 2x^3 - 3x^2 + x - 1", title="Função Polinomial")
```

### Propriedades da Função Polinomial
A função polinomial tem algumas propriedades importantes:

- A função é contínua em todos os pontos do domínio.
- A função é diferenciável em todos os pontos do domínio, e a derivada é sempre uma função polinomial de grau inferior.
- A função é integrável em todos os intervalos do domínio.

Essas propriedades fazem com que a função polinomial seja uma ferramenta útil em muitas aplicações matemáticas e científicas.

## Função Racional
Uma função racional é uma função que pode ser escrita na forma:

$$f(x) = \frac{p(x)}{q(x)}$$

onde:

$x$ é a variável de entrada
$p(x)$ e $q(x)$ são funções polinomiais
$q(x)$ não é igual a zero
### Exemplo Prático Manual
Vamos considerar a função racional $f(x) = \frac{x^2 + 1}{x + 1}$. Queremos calcular o valor da função para diferentes valores de $x$.

### Passo a Passo
Calcule o valor da função para $x = 0$:
$$f(0) = \frac{0^2 + 1}{0 + 1} = 1$$

Calcule o valor da função para $x = 1$:
$$f(1) = \frac{1^2 + 1}{1 + 1} = 1$$

Calcule o valor da função para $x = 2$:
$$f(2) = \frac{2^2 + 1}{2 + 1} = \frac{5}{3}$$

Observe que a função racional pode ter valores indefinidos quando o denominador é igual a zero.

>Resolução em Julia
Aqui está o código em Julia para definir e calcular a função racional:

```julia
# Defina a função racional
p(x) = x^2 + 1
q(x) = x + 1
f(x) = p(x) / q(x)

# Calcule o valor da função para diferentes valores de x
x_values = [0, 1, 2]
for x in x_values
    println("f($x) = $(f(x))")
end

# Plote a função
using Plots
x = -10:0.1:10
y = f.(x)
plot(x, y, label="f(x) = (x^2 + 1) / (x + 1)", title="Função Racional")
```

### Propriedades da Função Racional
A função racional tem algumas propriedades importantes:

- A função é contínua em todos os pontos do domínio, exceto quando o denominador é igual a zero.
- A função é diferenciável em todos os pontos do domínio, exceto quando o denominador é igual a zero.
- A função é integrável em todos os intervalos do domínio.

Essas propriedades fazem com que a função racional seja uma ferramenta útil em muitas aplicações matemáticas e científicas.

## Função Modular (ou Valor Absoluto)
A função modular, também conhecida como valor absoluto, é uma função que retorna o valor absoluto de um número real. Ela é definida como:

$$f(x) = |x|$$

onde:

$x$ é a variável de entrada
$|x|$ é o valor absoluto de $x$
### Exemplo Prático Manual
Vamos considerar a função modular $f(x) = |x|$. Queremos calcular o valor da função para diferentes valores de $x$.

### Passo a Passo
Calcule o valor da função para $x = 0$:
$$f(0) = |0| = 0$$

Calcule o valor da função para $x = 1$:
$$f(1) = |1| = 1$$

Calcule o valor da função para $x = -1$:
$$f(-1) = |-1| = 1$$

Calcule o valor da função para $x = 2$:
$$f(2) = |2| = 2$$

Calcule o valor da função para $x = -2$:
$$f(-2) = |-2| = 2$$

Observe que a função modular sempre retorna um valor não negativo.

> Resolução em Julia
Aqui está o código em Julia para definir e calcular a função modular:

```julia
# Defina a função modular
f(x) = abs(x)

# Calcule o valor da função para diferentes valores de x
x_values = [0, 1, -1, 2, -2]
for x in x_values
    println("f($x) = $(f(x))")
end

# Plote a função
using Plots
x = -10:0.1:10
y = f.(x)
plot(x, y, label="f(x) = |x|", title="Função Modular")
```

### Propriedades da Função Modular
A função modular tem algumas propriedades importantes:

- A função é contínua em todos os pontos do domínio.
- A função é diferenciável em todos os pontos do domínio, exceto no ponto $x = 0$.
- A função é integrável em todos os intervalos do domínio.

Essas propriedades fazem com que a função modular seja uma ferramenta útil em muitas aplicações matemáticas e científicas.

## Função Trigonométrica
Uma função trigonométrica é uma função que relaciona as razões entre os lados de um triângulo retângulo. As principais funções trigonométricas são:

- Seno (sin)
- Cosseno (cos)
- Tangente (tan)
- Cotangente (cot)
- Secante (sec)
- Cossecante (csc)
### Exemplo Prático Manual
Vamos considerar a função seno $f(x) = \sin(x)$. Queremos calcular o valor da função para diferentes valores de $x$.

### Passo a Passo

$$f(0) = \sin(0) = 0$$

Calcule o valor da função para $x = 30^\circ$:
$$f(30^\circ) = \sin(30^\circ) = \frac{1}{2}$$

Calcule o valor da função para $x = 60^\circ$:
$$f(60^\circ) = \sin(60^\circ) = \frac{\sqrt{3}}{2}$$

Calcule o valor da função para $x = 90^\circ$:
$$f(90^\circ) = \sin(90^\circ) = 1$$

Observe que a função seno é periódica com período $2\pi$.

> Resolução em Julia
Aqui está o código em Julia para definir e calcular a função seno:

```julia
# Defina a função seno
f(x) = sin(x)

# Calcule o valor da função para diferentes valores de x
x_values = [0, 30, 60, 90]
for x in x_values
    println("f($x) = $(f(x))")
end

# Plote a função
using Plots
x = 0:0.1:360
y = f.(x)
plot(x, y, label="f(x) = sin(x)", title="Função Seno")
```

### Propriedades da Função Trigonométrica
A função trigonométrica tem algumas propriedades importantes:

- A função é periódica com período $2\pi$.
- A função é contínua em todos os pontos do domínio.
- A função é diferenciável em todos os pontos do domínio.

Essas propriedades fazem com que a função trigonométrica seja uma ferramenta útil em muitas aplicações matemáticas e científicas.

## Função Irracional
Uma função irracional é uma função que não pode ser expressa como uma razão de polinômios, ou seja, não pode ser escrita na forma:

$$f(x) = \frac{p(x)}{q(x)}$$

onde $p(x)$ e $q(x)$ são polinômios.

### Exemplos de Funções Irracionais
Função raiz quadrada: $f(x) = \sqrt{x}$

Função raiz cúbica: $f(x) = \sqrt[3]{x}$

Função logarítmica: $f(x) = \log(x)$

Função exponencial: $f(x) = e^x$

### Propriedades das Funções Irracionais
As funções irrracionais têm algumas propriedades importantes:

- Não são polinômios
- Não são racionais
- Podem ser não contínuas ou não diferenciáveis em alguns pontos
- Podem ter comportamento irregular em alguns intervalos
### Exemplo Prático Manual
Vamos considerar a função raiz quadrada $f(x) = \sqrt{x}$. Queremos calcular o valor da função para diferentes valores de $x$.

### Passo a Passo
Calcule o valor da função para $x = 0$:
$$f(0) = \sqrt{0} = 0$$

Calcule o valor da função para $x = 1$:
$$f(1) = \sqrt{1} = 1$$

Calcule o valor da função para $x = 4$:
$$f(4) = \sqrt{4} = 2$$

Calcule o valor da função para $x = 9$:
$$f(9) = \sqrt{9} = 3$$

Observe que a função raiz quadrada é não negativa e cresce monotonamente.

> Resolução em Julia
Aqui está o código em Julia para definir e calcular a função raiz quadrada:

```julia
# Defina a função raiz quadrada
f(x) = sqrt(x)

# Calcule o valor da função para diferentes valores de x
x_values = [0, 1, 4, 9]
for x in x_values
    println("f($x) = $(f(x))")
end

# Plote a função
using Plots
x = 0:0.1:10
y = f.(x)
plot(x, y, label="f(x) = sqrt(x)", title="Função Raiz Quadrada")
```

Observe que o valor da função é sempre o mesmo, independentemente do valor de $x$.
