### **1ª Questão (Básica):**  
Quantos grupos de 2 pessoas podem ser formados a partir de 5 amigos?  

**Resposta:**  
- Combinação (ordem não importa):  
  $$C(5, 2) = \binom{5}{2} = \frac{5!}{2!3!} = \frac{5 \times 4}{2 \times 1} = \boxed{10}.$$  

---

### **2ª Questão:**  
De quantas maneiras diferentes 4 livros podem ser organizados em uma prateleira?  

**Resposta:**  
- Permutação simples (ordem importa):  
  $$P(4) = 4! = 4 \times 3 \times 2 \times 1 = \boxed{24}.$$  

---

### **3ª Questão:**  
Quantos anagramas diferentes podem ser formados com a palavra "SOL"?  

**Resposta:**  
- Letras distintas: permutação de 3 elementos:  
  $$3! = 6.$$  
  Anagramas: **SOL, SLO, OSL, OLS, LSO, LOS** → $\boxed{6}$.  

---

### **4ª Questão:**  
Quantos números de 3 dígitos distintos podem ser formados com {1, 2, 3, 4} se o número não pode começar com 1?  

**Resposta:**  
- Dígito inicial: 3 opções (2, 3, 4).  
- Restante: permutação de 2 dígitos:  
  $$3 \times P(3, 2) = 3 \times \frac{3!}{1!} = 3 \times 6 = \boxed{18}.$$  

---

### **5ª Questão:**  
De quantas maneiras 5 pessoas podem se sentar em uma mesa circular?  

**Resposta:**  
- Permutação circular:  
  $$(n-1)! = (5-1)! = 4! = \boxed{24}.$$  

---

### **6ª Questão:**  
Quantos anagramas da palavra "BANANA" têm todas as letras "A" juntas?  

**Resposta:**  
- Trate "AAA" como um único bloco:  
  Letras restantes: B, N, N, bloco → 4 elementos (com N repetido).  
  Permutação com repetição:  
  $$\frac{4!}{2!} = \frac{24}{2} = \boxed{12}.$$  

---

### **7ª Questão:**  
Uma sorveteria oferece 5 sabores. Quantos sorvetes de 3 bolas podem ser feitos se repetições são permitidas?  

**Resposta:**  
- Combinação com repetição:  
  $$C(n + k - 1, k) = \binom{5 + 3 - 1}{3} = \binom{7}{3} = \boxed{35}.$$  

---

### **8ª Questão:**  
Em um grupo de 8 pessoas, quantos comitês de 3 membros podem ser formados, incluindo um presidente e um vice?  

**Resposta:**  
1. Escolher presidente e vice (ordem importa):  
   $$P(8, 2) = 8 \times 7 = 56.$$  
2. Escolher o terceiro membro:  
   $$C(6, 1) = 6.$$  
3. Total:  
   $$56 \times 6 = \boxed{336}.$$  

---

### **9ª Questão:**  
Uma turma tem 10 homens e 7 mulheres. Quantos grupos de 5 pessoas têm exatamente 3 homens e 2 mulheres?  

**Resposta:**  
- Combinações separadas:  
  $$C(10, 3) \times C(7, 2) = \binom{10}{3} \times \binom{7}{2} = 120 \times 21 = \boxed{2520}.$$  

---

### **10ª Questão (Avançada):**  
Prove que:  
$$\binom{n}{k} = \binom{n-1}{k-1} + \binom{n-1}{k}.$$  

**Resposta:**  
- **Interpretação combinatória**:  
  - **Caso 1**: Um elemento específico está incluído no subconjunto:  
    Escolher $k-1$ elementos dos $n-1$ restantes: $\binom{n-1}{k-1}$.  
  - **Caso 2**: O elemento não está incluído:  
    Escolher $k$ elementos dos $n-1$ restantes: $\binom{n-1}{k}$.  
  - Total: $\boxed{\binom{n}{k} = \binom{n-1}{k-1} + \binom{n-1}{k}}$.

```julia
using Combinatorics

n = 5
k = 2
combinacoes = binomial(n, k)
println("Grupos de 2 pessoas: $combinacoes")
```

```julia
permutacoes = factorial(4)
println("Maneiras de organizar 4 livros: $permutacoes")
```

```julia
using Combinatorics

palavra = ["S", "O", "L"]
anagramas = collect(permutations(palavra))
println("Número de anagramas: $(length(anagramas))")
```

```julia
using Combinatorics

digitos = [1, 2, 3, 4]
numeros_validos = [join(num) for num in permutations(digitos, 3) if num[1] ≠ 1]
println("Números válidos: $(length(numeros_validos))")
```

```julia
presidente_vice = binomial(8, 2) * factorial(2)  # Escolha e ordem
terceiro_membro = binomial(6, 1)
total = presidente_vice * terceiro_membro
println("Comitês possíveis: $total")
```

```julia
homens = binomial(10, 3)
mulheres = binomial(7, 2)
grupos = homens * mulheres
println("Grupos com 3H e 2M: $grupos")
```

```julia
n = 5
k = 2
lado_esquerdo = binomial(n, k)
lado_direito = binomial(n-1, k-1) + binomial(n-1, k)
println("Prova: $lado_esquerdo = $lado_direito")
```

```julia
using Plots

# Dados para combinações
k_values = 1:5
combinacoes_homens = [binomial(10, k) for k in k_values]
combinacoes_mulheres = [binomial(7, k) for k in k_values]

plot(k_values, combinacoes_homens, label="Homens (n=10)", marker=:circle)
plot!(k_values, combinacoes_mulheres, label="Mulheres (n=7)", marker=:square)
xlabel!("k")
ylabel!("Combinações")
title!("Combinações por Tamanho de Grupo")
```
